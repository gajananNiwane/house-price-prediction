```python
# Lets try to understand which are important feature for the dataset


X=df.iloc[:,0:7] # independent  columns
print(X)

y=df.iloc[:,-1] # target column i.e price range
print(y)
df.head()
```

```python
# we want best feature target variable indentify we should make sure theat the target v ariable should be in values
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
y=np.round(df['price'])
# Apply SelectKBest class to extract top 5 best features

bestfeatures=SelectKBest(score_func=chi2,k=5)
fit=bestfeatures.fit(X,y)

dfscore=pd.DataFrame(fit.scores_)

dfcolumns=pd.DataFrame(X.columns)
# concat two dataframes for best visuliztion 


featureScores=pd.concat([dfcolumns,dfscore],axis=1)

featureScores.columns=['Specs','Score'] # nameing the dataframe columns
featureScores
```

```python
# All  the feature of the DataSet
```

```python
print(featureScores.nlargest(5,'Score'))
```

```python
# Feature Importance

from sklearn.ensemble import ExtraTreesClassifier

import matplotlib.pyplot as plt

model=ExtraTreesClassifier()
model.fit(X,y)
```
